## Country analysis with unsupervised Machine-Learning

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Country-data.csv")
```

```python
df1 = df.drop(columns="country")
```

```python
# Testa en en gpng
```

## Building the model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 5)
```

```python
df_clusters = kmeans.fit(df1)
```

```python
df_clusters.cluster_centers_
```

```python
df["Cluster"] = df_clusters.fit_predict(df1)
```

```python
df.Cluster.unique()
```
